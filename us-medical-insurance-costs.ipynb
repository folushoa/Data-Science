# U.S. Medical Insurance Costs

```python
#import csv library
import csv

#function to find most frequently occuring data
def data_frequency(data_dictionary):
    most_frequent_data = 0
    highest_frequency = 0
    
    for data in data_dictionary:
        if data_dictionary[data] > highest_frequency:
            most_frequent_data = data
            highest_frequency = data_dictionary[data]
    
    return most_frequent_data, highest_frequency

#function to count the number of times a values occurs
#store count in a dictionary, where the key is the value we are counting
def data_count(data_list):
    data_dictionary = {}
    
    for data in data_list:
        if data not in data_dictionary:
            data_dictionary[data] = 1
        else: 
            data_dictionary[data] += 1
    
    return data_dictionary

#put colums into list. Create empty list of columns
ages = []
sexes = []
bmis = []
number_of_children = []
smoker_status= []
regions = []
charges = []

#define function to load data from csv into a list
def load_list(csv_file, column_name, lst):
    #open csv file
    with open(csv_file) as csv_info:
        #read csv file
        csv_data = csv.DictReader(csv_info)
        #loop through each row of the csv file and save the data
        #in a list
        for row in csv_data:
            lst.append(row[column_name])
    
    #return list
    return lst

#load the csv data into its corresponding list
load_list('insurance.csv', 'age', ages)
load_list('insurance.csv', 'sex', sexes)
load_list('insurance.csv', 'bmi', bmis)
load_list('insurance.csv', 'children', number_of_children)
load_list('insurance.csv', 'smoker', smoker_status)
load_list('insurance.csv', 'region', regions)
load_list('insurance.csv', 'charges', charges)

#create dictionary
insurance_dictionary = {}
insurance_dictionary['Ages'] = [int(age) for age in ages]
insurance_dictionary['Sexes'] = sexes
insurance_dictionary['BMI'] = bmis
insurance_dictionary['Children'] = [int(children) for children in number_of_children]
insurance_dictionary['Smoker'] = smoker_status
insurance_dictionary['Region'] = regions
insurance_dictionary['Charges'] = [float(charge) for charge in charges]

#analysis using dictionary
#average of ages with insurance
total_age = 0
#total sum of all ages
for age in insurance_dictionary['Ages']:
    total_age += age

average_age = round(total_age/len(insurance_dictionary['Ages']), 0)

print("The average age of people with insurance is {average_age}".format(average_age = average_age))

#frequency of ages. Here we create a dictionary to count and store the freuquency of each age
age_frequency = data_count(insurance_dictionary['Ages'])
        
#Find the age that occurs the most
most_frequent_age ,highest_frequency = data_frequency(age_frequency)

print("Most people signed up for insurance are of the age {most_frequent_age}".format(most_frequent_age = most_frequent_age))

#Region with the most insured
#create a dictionary to hold the count of the region
region_frequency = data_count(insurance_dictionary['Region'])

#find the region with the highest frequency
most_frequent_region, highest_region_frequency = data_frequency(region_frequency)

print("Most people signed up for insurance from the {most_frequent_region} region".
      format(most_frequent_region = most_frequent_region))

#find the total charge of smokers to non-smokers and compare
smokers_total_charge = 0
non_smokers_total_charge = 0
for row in range(len(insurance_dictionary['Charges'])):
    if insurance_dictionary['Smoker'][row] == 'yes':
        smokers_total_charge += insurance_dictionary['Charges'][row]
    elif insurance_dictionary['Smoker'][row] == 'no':
        non_smokers_total_charge += insurance_dictionary['Charges'][row]

difference = round(abs(smokers_total_charge - non_smokers_total_charge), 2)

if smokers_total_charge > non_smokers_total_charge:
    print('Smokers where charged more than non smokers by $' + str(difference))
else:
    print('Non-smokers where charged more than smokers by $' + str(difference))
```
